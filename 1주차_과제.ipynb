### **언어**
* **네이버 파파고(Papago) 번역 서비스.**

 * 개요
  
  국내 독보적인 포털 사이트로서 방대한 데이터를 보유한 네이버가 개발하고 제공 중이며 현재 **국내 가장 인지도 있는 번역 서비스**이다.

 * 기반 기술
 
  N2MT (Naver Neural Machine Translation)라고 불리는 인공신경망(Artificial Neural Network) 기술에 기반하였으며 NAVER LABS에서 자체 개발하였다. 기존 SMT(Statistic Machine Translation)에 일부 딥러닝을 적용한 방법과는 달리 **전체 번역 프레임워크를 딥러닝 방식으로 구축해 번역 정확도를 크게 향상**하였다고 한다.

 * 지원 언어

 한국어, 일본어, 중국어, 영어, 스페인어, 프랑스어, 베트남어, 태국어, 인도네시아어, 독일어, 러시아어, 포르투갈어, 힌디어 13개의 언어를 번역할 수 있다. 

 * 번역 특징
 
 단순히 개별 단어의 번역이 아니라, **문장 내 요소 관계와 맥락에 맞게 번역**된다. 문장을 입력해보면, 단어들이 입력되는 과정에서 이미 출력된 번역도 뒤에 단어들이 추가 입력되면서 맥락을 고려하여 번역 결과가 달라진다. 예를 들어 "This homework seems a bit hard."라는 문장을 입력한다면, 특히 동사인 seems 다음을 입력하는 시점에 단어가 추가될 때마다 그에 맞게 앞 번역 결과가 역동적으로 바뀐다. 그 변화는 의미의 정확도뿐만 아니라, 표현의 자연스러움을 이룬다.

 * 훈련 정책: 유저 평가 수집, GYM, 우수 트레이너 제도
 
  네이버는 **유저들의 번역물 평가 피드백을 통해 품질을 개선**하고 있다. 파파고 인터페이스에 만족과 불만족 버튼이 있어 유저가 번역 결과 평가를 줄 수 있다. 심지어 [PAPAGO GYM](https://papago.naver.com/gym/home)이라는, 유저들의 피드백 수집 전용 서비스에서는 10개의 번역 전 문장과 번역 후 문장의 쌍을 제시하여 평가를 하게 하고 그 대가로 파파고 포인트와 네이버페이 포인트를 지급한다. 유저들이 '좋아요'와 '어색해요' 중 하나를 고를 수 있을 뿐만 아니라, '어색해요'를 고르는 경우 어떤 이유에 해당하는지 의미 생략, 불필요한 반복, 오역, 문법 오류, 기타 항목 중 선택하고, 더 나은 번역 표현도 제시할 수 있다.
 
  심지어 앞으로 '우수 트레이너' 제도를 도입하여 정확한 피드백으로 많이 기여하는 유저들에게 더 큰 혜택을 주려고 하는 등 파파고 엔진 훈련에 필요한 유저의 평가 정보를 수집하는 유인책으로 활용하려 하고 있다.

 * 연관 기술

 이미지에 있는 텍스트를 번역해주는 기능, 음성 입력을 번역해주는 기능 등이 앱에 통합되어 있어, 음성, 이미지 관련 인공지능 기술도 연계된다.

 * 나의 생각

 번역 서비스처럼 대량의 데이터를 필요로 하는 Ai는 역시 **양질의 대량의 데이터를 수집하는데 유리한 플랫폼을 쥔 기업이 매우 유리한 것 같다.** 인터넷 검색부터가 텍스트에 기반하며, 글로벌 시대에 외국어 텍스트도 검색 포털에서 취급하는 주요한 
데이터가 되어 가고 있는 것 같다.

 게다가 네이버는 네이버쇼핑 및 네이버페이라는 금전적인 보상과 연계할 수 있는 플랫폼까지 쥐고 있다. 위 훈련 정책 항목에서 언급했듯이 **딥러닝 신경망을 개선하기 위한 양질의 데이터를, 네이버페이 포인트 등의 경제적 보상을 통해 손쉽게 이끌어낼 수 있다.**

 미래 AI 시장이 네이버와 같은, 플랫폼을 장악한 큰 기업들에게 종속될 수 있는데, 다양한 아이디어의 혁신이 이루어지려면 작은 스타트업 기업들에게도 기회가 주어져야 한다. 정부가 이미 공공 데이터를 개방하여 온 것은 긍정적이지만 그것으로는 불충분한 것 같다.
 
  플랫폼 기반이 없는 스타트업 기업들도, 공공에 가까운 민간 플랫폼의 데이터를 작은 비용으로 이용하게 하는 등 규제 및 지원 정책이 있어야, 현재 과제를 수행 중인 광주 인공지능사관학교가 목표하는 광주 인공지능 생태계 형성도 순탄할 것이라 생각한다.

### **음성**
* **클로바 보이스**
 * 개요

 이 역시 국내 최대 포털인 네이버에서 제공하는 음성 합성 기술로서 **인공지능 플랫폼 클로바**의 음성 합성 부문 기술 서비스이다. 구체적으로, 인공지능 스피커와의 대화에 필요한 음성 관련 기술 부문에는 음성인식, 음성합성, 자연어처리(NLP)가 있는데, 이 중 **음성합성** 부문을 포괄한다.

 * 음성합성 기술 세계적 배경.

  미국엔 뉘앙스커뮤니케이션, 중국엔 아이플라이텍(iFLYTEK), 유럽엔 이보나(IVONA Software), 국내엔 네이버가 독보적이라고 한다. 중국 바이두도 수천명의 인력을 두고 기술 개발에 매진하고 있다. 이보나는 미국 아마존에 인수됐다. 애플은 AI 비서 시리에 미국 뉘앙스 기술을 활용했다. 구글은 자체 음성합성팀을 운영 중으로 구글이 음성합성 분야 최강자로 평가된다고 한다.

 * 활용 현황

  클로바 보이스의 **머신러닝** 기술은 기본적으로 네이버가 서비스 중인 클로바 스마트 스피커, 네비게이션 앱, 뉴스, 파파고, 오디오북 등 음성 합성이 필요한 영역에 크게 활용되고 있다. 그 외 일반 사용자 및 기업 사용자들이 활용하기 위한 서비스로 다음이 있다. 텍스트를 음성으로 변환하여 동영상에 더빙하는 클로바 더빙, 네이버 클라우드에서 유료로 제공하는 API 서비스인 CSS(Clova Speech Synthesis), Clova Premium Voice(CPV)이 있다.

 * 기반 기술

   * HDTS(High-quality DNN Text-to-Speech)
   
   HDTS는 **특정 분야에 필요한 목소리 특성**을 살려 최고 품질의 합성음을 만드는 기법으로서 클로바 스마트 스피커와 오디오북의 유인나 목소리를 제작하는데 쓰였다. 
   
   * NES(Natural End-to-End Speech Synthesis)

   **짧은 녹음만으로 고품질의 합성음을 제작할 수 있는 신기술**이다. 40분 수준의 음성 데이터(400문장)로 실제 사람의 음성에 가까운 자연스러운 합성음을 제작할 수 있는 기술로, 40시간에서 100시간 정도의 음성이 필요했던 기존과 비교하여 쉽고 빠르게 음성 데이터를 만들 수 있다. 특정 분야에 대한 제약 없이 사용할 수 있으며, 기본 목소리 외에 **기쁨·슬픔 등 감정이 반영된 목소리도 제공**한다.

 * 나의 생각

  **인간이 가장 편안하게 사용하는 소통 수단이며, 다양한 감정을 효과적으로 전달할 수 있는 수단이 음성이므로, 음성 기술은 인간이 장차 활용할 인공지능 기술 중 핵심 영역이 될 것 같다.** 예를 들어, 가정 내 또는 차량 내처럼 정서적 상호작용이 중요한 곳에서 어떤 기계적 성능을 넘어 인간의 마음을 어루만져 주는 역할까지도 할 수 있을 것이다.
   
  게다가, **발성에 장애가 있는 사람들에게는 삶의 질을 확 바꾸는 보조 도구가 될 수 있다.** 그런데 우리나라는 음성 합성 전공자도 드물고 관련 연구 기반도 부족하다고 한다. 음성 기술은 공공의 이익에 기여할 가능성이 높으므로, 네이버 같은 사기업이 주도하는 것 외에도, 국가적 지원을 통해 교육계의 연구 기반도 확충하는 것도 필요할 것 같다.
  
  조사를 수행하는 과정에서 음성 인공지능 기술에 대한 관심이 커졌는데, **특히 감정의 톤을 담는 것, 단 40분만의 음성 데이터만으로 음성 합성을 할 수 있는 것 등이 매우 신기하여 앞으로 가능하다면 공부를 해보고 싶다.**

   네이버 보이스 팀이 NES 같은 혁신적인 기술 수준까지 산출할 수 있었던 것은 팀 내 소통이 원활한 분위기가 한몫했다는 뉘앙스의 인터뷰 기사를 봤는데, 모호하고 다양한 영역이 혼재된 **인공지능 기술의 영역에서는 팀원 간 소통과 협력이 중요하겠다는 생각이 들었다.**

### **이미지 인식**
* **구글 렌즈**
 * 개요

 구글이 개발한, 이미지에 있는 사물을 파악하는 기술이다. 안드로이드 표준 카메라 앱에 통합되어 있고 구글 어시스턴트, 구글 포토에서도 이용 가능하다.

 * 기반 기술

  구글이 개발한 **딥러닝 인공지능 신경망에 기반**한다. 사물을 인식하고 관련된 정보를 추출할 수 있는 것 외에도 번역, 검색 등 여러 구글 기술과 서비스와 연동되어 **다양하게 활용**할 수 있게 되어 있다.

  * 나의 생각

 스마트폰을 통해 디지털 라이프를 영위하면서 가장 자주 쓰는 것 중 하나가 사진 촬영, 이미지 공유 등 이미지 관련한 기능일 것이다. **구글이 세계 최고 점유율의 검색 엔진과 안드로이드 플랫폼을 보유하면서 방대한 데이터와 서비스를 가지고 있으므로 이미지 인식 인공지능 기술 및 제품 개발에 매우 유리한 것 같다.** 거기다가 구글은 이미 인공지능 전반에 거쳐 전세계 최고 수준의 인재와 연구 결과를 가지고 있다.

 그 인공지능을 이용한 흥미로운 예시로는, 와이파이 공유기의 SSID 스티커를 촬영하면 자동으로 해당 와이파이에 연결시켜주거나, 고양이를 촬영하면 고양이가 어떤 종인지를 분석해주거나, 어떤 유명 미술 작품을 촬영하면 어떤 작품인지 찾아주는 등이 가능한 것이다. 앞으로 구글 렌즈 기술은 휴대폰을 통한 이용을 넘어, CCTV나, 매장 관리, 교통 상황 파악, 자율 주행 자동차 등에 응용될 수 있다.

  다만 우려되는 것은 이미지 인식 기술의 경우, 사생활을 침해하는 데에 사용될 수 있다는 것이다. 특히 0.1초의 얼굴 노출만으로 신원을 특정할 수 있게 된다면, 특히 중국처럼 국가를 권위로 통치하는 사회 시스템의 경우, 인민 또는 국민 감시 수단으로 활용될 수 있을 것 같다. **인식된 정보들의 악용 방지를 위한 기술과 제도의 발전도 필요할 것 같다.**

### **자율주행**

* TESLA의 자율주행 기술: **오토파일럿**

 * 장착 센서

 신차 기준으로 전방을 탐지하는 **레이더 센서 1개**, **초음파센서 12개**, 거리에 따른 **전방 카메라 3대, 후방 1개, 전우측, 전좌측, 후우측, 후좌측 각각 1개씩 총 8개의 카메라**가 장착되었다.

 * 기술

  자율주행은 미국 미국자동차 기술자 협회(SAE) 기준으로 레벨0(無자율주행), 레벨1(운전자 지원), 레벨2(부분 자율주행), 레벨3(조건부 자율주행), 레벨4(고도 자율주행), 레벨5(완전 자율주행) 등으로 구분되는데, 테슬라의 오토파일럿은 **레벨 2 ~ 레벨 3 수준**으로 평가받고 있으며, 장기적으론 레벨 5 완전 자율주행을 목적으로 발전하고 있다.

  오토파일럿은 차선 내에서 차량을 자동으로 조항, 가속, 제동을 해 준다. 네비게이션과 함께 구동하여 앞차와 속도 제한에 맞게 속도도 조절하고 차선도 변경하며 고속도로 교차로 및 출구로 자동 조향하여 목적지로 운전해준다. 그러나, 도로 상황이 복잡하는 등 오토파일럿의 능력을 벗어난 상황에서는 활성화가 되지 않거나 취소가 된다.

 자율주행은 **이미지 처리와 깊게 관계**된다. 앞 장착 센서 항목에서 언급했던 센서들에서 입력한 시각 데이터를 기반으로 교통 신호, 보행자, 다른 차, 차선 등 주위 사물을 파악하게 된다. 이를 **딥러닝 신경망으로 구현한 인공지능 기술을 테슬라 비전**이라 한다.

 무엇보다도, **테슬라가 자율주행 인공지능 개발에서 누구보다도 앞선 점은 실시간으로 현실의 주행 데이터를 초대량으로 수집하고 있다는 점**이다.
 앞에서 언급했던 다른 분야의 인공지능 기술들을 보유했던 네이버, 구글도 인공지능 기술 기반이 그들이 가진 대량의 데이터였다.

 그런데 그 구글의 웨이모도 자율주행 분야에서 데이터 수집량을 보면 테슬라에게 비교가 되지 않는다. 웨이모가 금년 2020년 초 기준으로 데이터 수집용 차량 1000여대, 실제 도로 주행거리가 3천6백만km인 반면, **테슬라는 70만대의 차량으로부터 하루면 웨이모의 전체 누적 데이터량만큼 수집**할 수 있다고 한다. 총 35억km의 오토파일럿 실제 주행거리를 가지고 있다. 데이터를 통한 머신 러닝에 기반하고 있으므로, 테슬라가 얼마나 유리한 고지를 가지고 있는지 알 수 있다. 나중에 우리 현대 자동차 같은 후발주자들은 어쩌면 테슬라에게 돈을 주고 데이터를 사야 하는 상황이 올 수도 있을 것 같다.

 * 테슬라가 말하는 **완전자율주행의 미래**

 아래는 테슬라 홈페이지에서 발췌한 것으로 그들이 목표한, 완전 자율 주행의 미래를 묘사하고 있다. 운전자가 음성 인식을 통한 목적지만 입력하면 되며, 심지어 미리 입력된 일정에 따라 자동으로 목적지를 설정하여 운행하고 돌아올 수 있다는 것이다. 다양한 신호 규제가 있어 사람에게도 복잡한 상황의 도로에서도 자율 주행이 되는 것을 목표로 하고 있다.
> _"차에 타서 목적지를 말하기만 하면 됩니다. 아무 말도 하지 않으면 자동차는
달력을 확인해서 가정한 목적지로 가거나, 달력에 일정이 따로 없으면 집으로 갑니다.
Tesla는 최적의 경로를 계산해서 도시 거리 (차선 표시가 없어도)를 주행하며,
신호등, 정지 신호 및 로터리가 있는 복잡한 교차로도 주행할 수 있고,
많은 차량들이 빠르게 이동하는 밀집되고 혼잡한 고속도로에서도 주행할 수 있습니다.
목적지에 도착해서 입구에 내리면 차량이 주차 탐색 모드로 진입한 후,
자동으로 주차 공간을 찾아서 주차합니다. 핸드폰을 탭해서 다시 차량을 부를 수 있습니다."_

 * 나의 생각

 원래는, 자율 주행을 위해 설계된 도로 시스템의 도움이 필요없이 기존의 모든 도로를 알아서 주행해주는 것이 진정한 자립형 인공지능일 것이다. 그러나 테슬라 홈페이지에서 말하는 완전자율주행의 목표는, 수많은 예외 상황이 존재하며 인명이 손상될 위험이 존재하는 도로 상황에서는 아주 먼 목표가 될 것 같다. 물론 그런 이상적 목표를 갖고 끊임없이 딥러닝을 기반으로 인공지능을 개선해 나가는 것은 타당하다.

 내 생각은, **도로 시스템 자체에 자율주행을 돕는 설계를 도입하고 발전시켜 간다면, 자율주행 기능의 혜택을 인간 문명이 더 신속히 활용할 수 있을 것 같다.** 이는 공공의 일이므로, 정부 차원의 추진, 도로 교통법의 변화 등이 수반되야 한다. 그렇게 되기 위해서는 테슬라나 구글, 현대 자동차 같은 기술 기업들의 연구 외에, **도로 연구나 법 연구 등과 관련된 교육 부분부터 딥러닝 이해 같은 부분이 연계되어야 할 것 같다.**
 
  여기서 일반화하여 깨달은 바는 다음과 같다. 도로 시스템, 교통법 외에도 수많은 사회 규약과 인프라가 스마트화되려면 각 분야의 종사자들이 인공지능의 기본 요건이라도 이해해야 한다. **인공지능 기술을 공부하여 일정의 소양을 갖춘다면, 당장 설령 인공지능 전문 기업이 아닌 분야에 종사하더라도, 인공지능의 큰 물결에 호응하고 국가에 기여할 수 있는 더 나은 인재가 될 수 있다고 생각한다.** 나아가, 선진국에서 코딩을 의무 교육으로 도입하는 추세인 현재, 인공지능 교육이 보급될수록 그 나라가 발전할 잠재력이 커질 것이라 믿게 되었다.
